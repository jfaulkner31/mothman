```python
import HeatExchanger.HX as HX
import numpy as np

area = 1
htc = 125000.0
temp_rel_tol = 1e-10
hot_bc =  657.0+273.15
cold_bc = 579.0+273.15

hot_dict = {
  'pressure': 101325.0,
  'mdot': 162,
  'cp': 1966,
  'mass': 348
}

cold_dict = {
  'pressure': 101325.0,
  'mdot': 100,
  'cp': 2390,
  'mass': 100
}

# Make object
the_hx = HX.HeatExchangerObject(area=area, htc=htc, hot_dict=hot_dict, cold_dict=cold_dict, temp_rel_tol=temp_rel_tol)

# Set hot and cold bc
the_hx.set_hot_bc(T_in=hot_bc, mdot=hot_dict['mdot'], pressure=hot_dict['pressure'])
the_hx.set_cold_bc(T_in=cold_bc, mdot=cold_dict['mdot'], pressure=cold_dict['pressure'])
the_hx.initialize_outlets(transferred_heat=8e6)

# Solve something.
the_hx.solve(_dt=1e321, printSolve=True)
the_hx.update_old_to_most_recent()
the_hx.print_solution()
print(the_hx.get_heat_transferred())

timesteps = np.linspace(0,300,3000)
dts = timesteps[1:] - timesteps[0:-1]
# t = 0.0
# for dt in dts:
#   t+=dt
#   print(f"Time = {t}")
#   the_hx.solve(_dt=dt, printSolve=True)
#   the_hx.update_old_to_most_recent()
```
